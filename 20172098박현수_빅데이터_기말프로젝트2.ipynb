# 지역별 피해

```python
df_gangwon[df_gangwon['tm'].dt.year==2020]['tm']
```

```python
# 강원도 일별 강수량
plt.figure(figsize=(14,7))
plt.plot(df_gangwon[df_gangwon['tm'].dt.year==2020]['tm'],df_gangwon[df_gangwon['tm'].dt.year==2020]['avgTs'],'-o',label='평균지면온도')
plt.plot(df_gangwon[df_gangwon['tm'].dt.year==2020]['tm'],df_gangwon[df_gangwon['tm'].dt.year==2020]['sumRn'],'-^',label='일별강수')
plt.plot(df_gangwon[df_gangwon['tm'].dt.year==2020]['tm'],df_gangwon[df_gangwon['tm'].dt.year==2020]['maxTa'],'-*',label='최고기온')

plt.xticks(rotation=45,fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
plt.title("강원도",size=30)
plt.legend(fontsize=20)
plt.show()
```
